## Investigation Overview

> This is a close look at how various attributes affect the amount of loan a inndividual can access and the APR of such loan. The main focus circles around the original loan amount, the prosper rating, Borrower APR and Income rate

## Dataset Overview

> The original data from prosper includes 113937 rows of data with 87 variables. Variables with less than 50% valuable data was cleared after wrangling of the data making the total data consist of 84853 rows annd 70 variables.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_data = pd.read_csv("loan_data.csv");
```

## Loan Original Amount Distribution 

> This te the distribution of the original loan amount accessed by borrowers according to the data from prosper. More borrowers have accessed lesser value from 25k annd below.

```python
bins = np.arange(0, loan_data["LoanOriginalAmount"].max()+1000, 1000)
plt.figure(figsize=[10,6])
plt.hist(data = loan_data, x = "LoanOriginalAmount", bins = bins)
plt.xlabel("Loan Original Amount")
plt.ylabel("count")
plt.title("Loan Original Amount distribution")
plt.show()
```

## BorrowAPR against the Loan Original Amount

> Although, the APR has a broad range for a lesser amount of loan. The amount of interest decreases with increase in loan amount.

```python
plt.figure(figsize = [8, 6])
sb.regplot(data = loan_data, x = 'LoanOriginalAmount', y = 'BorrowerAPR', scatter_kws={'alpha':0.01});
plt.xlabel('Loan Amount ($)')
plt.ylabel('Borrower APR')
plt.title('Borrower APR vs. Loan Amount');
```

## Relationship between BorrowerAPR, LoanOriginalAmout annd ProsperRating

> This shows a clearer relationship between how Loan amount is affected by the prosper rating. It also shows how the loan amount affects the interest rate of the borrower.

```python
sb.relplot(data=loan_data.sample(500, random_state=1), y='BorrowerAPR', x='LoanOriginalAmount', 
            hue='ProsperRating (Alpha)', palette='PuBu', height=4, aspect=1.5);
```
